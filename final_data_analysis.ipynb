# Final Data Analysis

```python
# Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
from scipy.stats import sem
from scipy.stats import linregress
```

```python
# Set path and read CSV
path = 'ocean_data/MERGE.csv'
merged_df = pd.read_csv(path)
merged_df
```

# Let's start plotting!
## Total Pop and World Happiness Score

```python
# 2015 Total Pop and 2015 World Happiness

total_2015 = merged_df.iloc[:, 1]
whs_2015 = merged_df.iloc[:, 58]
total_whs_2015 = st.pearsonr(total_2015, whs_2015)
print(f'The correlation between total population and world happiness score in 2015 is {round(total_whs_2015[0],2)}.')

x_total_2015 = merged_df['2015 Happiness Score']
y_total_2015 = merged_df['2015 Total Pop']
slope, intercept, rvalue, pvalue, stderr = linregress(x_total_2015, y_total_2015)
regressValues = x_total_2015 * slope + intercept
lineEq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

plt.scatter(x_total_2015, y_total_2015)
plt.plot(x_total_2015, regressValues,"r-")
plt.annotate(lineEq,(6,10),fontsize=15,color="red")
plt.xlabel('2015 Happiness Score')
plt.ylabel('2015 Total Population')
plt.title('2015 Happiness Scores and Total Population')
plt.show()
```
