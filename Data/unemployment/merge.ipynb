```python
# Dependencies
import pandas as pd

# Set up file paths
new_variable_path = '../ocean_data/new_variable_merge.csv'
gdp_path = '../project_data_teare/Output/GDP_API_clean.csv'
labor_path = '../project_data_teare/Output/Labor_API_clean.csv'
gov_expend_path = '../project_data_petrus/ed_spending.csv'
literacy_path = '../data_pmprybylski/Output/adult_literacy.csv'
whs_path = '../data_pmprybylski/Output/whs_combined.csv'
```

```python
# Import population and unemployment CSVs
new_variable_df = pd.read_csv(new_variable_path)
new_variable_df = new_variable_df.drop(columns=['Unnamed: 0'])
new_variable_df.head()
```

```python
# Import GDP CSV
gdp_df = pd.read_csv(gdp_path)
gdp_df = gdp_df.drop(columns=['Unnamed: 0'])
gdp_df.head()
```

```python
# Import Labor CSV
labor_df = pd.read_csv(labor_path)
labor_df = labor_df.drop(columns=['Unnamed: 0'])
labor_df.head()
```

```python
# Import world happiness CSV
whs_df = pd.read_csv(whs_path)
whs_df
```

```python
# Merge pop/unemployment and GDP Dataframes
new_variable_df = new_variable_df.merge(gdp_df, how='left', on='Country Name')
new_variable_df.head()
```

```python
# Merge pop/unemployment/GDP and labor Dataframes
new_variable_df = new_variable_df.merge(labor_df, how='left', on='Country Name')
new_variable_df.head()
```

```python
# Inner join pop/etc and WHS Dataframes
merged_inner_df = pd.merge(left=new_variable_df, right=whs_df, left_on='Country Name', right_on='Country Name')
merged_inner_df
```

```python
# Drop NaN values
merged_inner_df = merged_inner_df.dropna()
merged_inner_df
```

```python
merged_inner_df.to_csv('MERGE.csv', index=False)
```
