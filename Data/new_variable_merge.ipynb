```python
# Dependencies
import pandas as pd
import matplotlib.pyplot as plt

# Set file paths
adult_pop_path = "population/adult_pop.api.csv"
total_pop_path = "population/total_pop_api.csv"
unemployment_female_path = "unemployment/unemployment_female/unemployment_female_api.csv"
unemployment_male_path = "unemployment/unemployment_male/unemployment_male_api.csv"
unemployment_total_path = "unemployment/unemployment_total/unemployment_total_api.csv"
electricty_path = "access_to_electricity/access_to_electricity.csv"
infant_mortality_path = "infant_mortality_rate/infant_mortality_rate.csv"
life_expect_female_path = "life_expectancy/life_expectancy_female.csv"
life_expect_male_path = "life_expectancy/life_expectancy_male.csv"
military_expend_path = "life_expectancy/military_expenditure.csv"
```

## Import CSVs

```python
# Import adult (15-64) population csv, skip blank rows, and select the data we want
adult_pop_df = pd.read_csv(adult_pop_path, skiprows=4)
adult_pop_df = adult_pop_df[['Country Name', '2015', '2016', '2017', '2018', '2019']]
adult_pop_df.head()
```

```python
# Import total population csv, skip blank rows, and select the data we want
total_pop_df = pd.read_csv(total_pop_path, skiprows=4)
total_pop_df = total_pop_df[['Country Name', '2015', '2016', '2017', '2018', '2019']]
total_pop_df.head()
```

```python
# Import female unemployment csv, skip blank rows, and select the data we want
unemployment_female_df = pd.read_csv(unemployment_female_path, skiprows=4)
unemployment_female_df = unemployment_female_df[['Country Name', '2015', '2016', '2017', '2018', '2019']]
unemployment_female_df.head()
```

```python
# Import male unemployment csv, skip blank rows, and select the data we want
unemployment_male_df = pd.read_csv(unemployment_male_path, skiprows=4)
unemployment_male_df = unemployment_male_df[['Country Name', '2015', '2016', '2017', '2018', '2019']]
unemployment_male_df.head()
```

```python
# Import total unemployment csv, skip blank rows, and select the data we want
unemployment_total_df = pd.read_csv(unemployment_total_path, skiprows=4)
unemployment_total_df = unemployment_total_df[['Country Name', '2015', '2016', '2017', '2018', '2019']]
unemployment_total_df.head()
```

```python
# Import access to electricity csv, skip blank rows, and select the data we want
electricity_df = pd.read_csv(electricty_path, skiprows=4)
electricity_df = electricity_df[['Country Name', '2015', '2016', '2017', '2018']]
electricity_df.head()
```

```python
# Import infant mortality csv, skip blank rows, and select the data we want
infant_mortality_df = pd.read_csv(infant_mortality_path, skiprows=4)
infant_mortality_df = infant_mortality_df[['Country Name', '2015', '2016', '2017', '2018', '2019']]
infant_mortality_df.head()
```

```python
# Import life expectancy female csv, skip blank rows, and select the data we want
life_expect_female_df = pd.read_csv(life_expect_female_path, skiprows=4)
life_expect_female_df = life_expect_female_df[['Country Name', '2015', '2016', '2017', '2018']]
life_expect_female_df.head()
```

```python
# Import life expectancy male csv, skip blank rows, and select the data we want
life_expect_male_df = pd.read_csv(life_expect_male_path, skiprows=4)
life_expect_male_df = life_expect_male_df[['Country Name', '2015', '2016', '2017', '2018']]
life_expect_female_df.head()
```

```python
# Import military expenditure csv, skip blank rows, and select the data we want
military_expend_df = pd.read_csv(military_expend_path, skiprows=4)
military_expend_df = military_expend_df[['Country Name', '2015', '2016', '2017', '2018', '2019']]
military_expend_df.head()
```

## Merge Data

```python
# Merge the two population dataframes (adult and total) together
population_df = total_pop_df.merge(adult_pop_df, how='left', on='Country Name')
population_df.head()
```

```python
# Rename the columns
population_df = population_df.rename(columns={'2015_x':'2015 Total Pop', '2016_x':'2016 Total Pop', '2017_x':'2017 Total Pop',
                                              '2018_x':'2018 Total Pop', '2019_x':'2019 Total Pop',
                                              '2015_y':'2015 Adults (%)', '2016_y':'2016 Adults (%)', '2017_y':'2017 Adults (%)',
                                              '2018_y':'2018 Adults (%)', '2019_y':'2019 Adults (%)'})
population_df.head()
```

```python
# Merge the total and female unemployment rate dataframes together
unemployment_df = unemployment_total_df.merge(unemployment_female_df, how='left', on='Country Name')
unemployment_df.head()
```

```python
# Rename the columns 
unemployment_df = unemployment_df.rename(columns={'2015_x':'2015 Total Unemployment', '2016_x':'2016 Total Unemployment',
                                                  '2017_x':'2017 Total Unemployment', '2018_x':'2018 Total Unemployment', 
                                                  '2019_x':'2019 Total Unemployment',
                                                  '2015_y':'2015 Female Unemployment', '2016_y':'2016 Female Unemployment', 
                                                  '2017_y':'2017 Female Unemployment', '2018_y':'2018 Female Unemployment', 
                                                  '2019_y':'2019 Female Unemployment'})
unemployment_df.head()
```

```python
# Merge the unemployment and male unemployment dataframes together
unemployment_df = unemployment_df.merge(unemployment_male_df, how='left', on='Country Name')
unemployment_df.head()
```

```python
# Rename the new columns
unemployment_df = unemployment_df.rename(columns={'2015': '2015 Male Unemployment', '2016': '2016 Male Unemployment', 
                                                  '2017': '2017 Male Unemployment', '2018': '2018 Male Unemployment', 
                                                  '2019': '2019 Male Unemployment'})
unemployment_df.head()
```

```python
# Merge the population and unemployment dataframes together
pop_unemployment_df = population_df.merge(unemployment_df, how='left', on='Country Name')
pop_unemployment_df.head()
```

```python
# Merge the pop/unemployment and electricity access DataFrames together
merged_df = pop_unemployment_df.merge(electricity_df, how='left', on='Country Name')
merged_df.head()
```

```python
# Rename the new columns
merged_df = merged_df.rename(columns={'2015': '2015 Access to Electricity', '2016': '2016 Access to Electricity', 
                                                  '2017': '2017 Access to Electricity', '2018': '2018 Access to Electricity', 
                                                  '2019': '2019 Access to Electricity'})
merged_df.head()
```

```python
# Merge the pop/etc. and infant mortality DataFrames together
merged_df = merged_df.merge(infant_mortality_df, how='left', on='Country Name')
merged_df.head()
```

```python
# Rename the new columns
merged_df = merged_df.rename(columns={'2015': '2015 Mortality (<5)', '2016': '2016 Mortality (<5)', 
                                                  '2017': '2017 Mortality (<5)', '2018': '2018 Mortality (<5)', 
                                                  '2019': '2019 Mortality (<5)'})
merged_df.head()
```

```python
# Merge the pop/etc. and life expectancy female DataFrames together
merged_df = merged_df.merge(life_expect_female_df, how='left', on='Country Name')
merged_df.head()
```

```python
# Rename the new columns
merged_df = merged_df.rename(columns={'2015': '2015 Female Life Expectancy', '2016': '2016 Female Life Expectancy', 
                                                  '2017': '2017 Female Life Expectancy', '2018': ' 2018 Female Life Expectancy'})
merged_df.head()
```

```python
# Merge the pop/etc. and life expectancy male DataFrames together
merged_df = merged_df.merge(life_expect_male_df, how='left', on='Country Name')
merged_df.head()
```

```python
# Rename the new columns
merged_df = merged_df.rename(columns={'2015': '2015 Male Life Expectancy', '2016': '2016 Male Life Expectancy', 
                                                  '2017': '2017 Male Life Expectancy', '2018': ' 2018 Male Life Expectancy'})
merged_df.head()
```

```python
# Merge the pop/etc. and military expenditure DataFrames together
merged_df = merged_df.merge(military_expend_df, how='left', on='Country Name')
merged_df.head()
```

```python
# Rename the new columns
merged_df = merged_df.rename(columns={'2015': '2015 Military Expenditure (%GDP)', '2016': '2016 Military Expenditure (%GDP)', 
                                      '2017': '2017 Military Expenditure (%GDP)', '2018': ' 2018 Military Expenditure (%GDP)',
                                      '2019': '2019 Military Expenditure (%GDP)'})
merged_df.head()
```

## Print the DF

```python
# Print the dataframe to an Excel sheet
merged_df.to_csv('new_variable_merge.csv')
```
