```python
import pandas as pd
```

```python
gdp_file = "Data Files/GDP_API.csv"
```

```python
gdp_file_df = pd.read_csv(gdp_file, skiprows=4)
gdp_file_df.head()
```

```python
gdp_file_df.columns
```

```python
reduced_df = gdp_file_df[['Country Name', '2015', '2016', '2017', '2018', '2019']]
reduced_df.head()
```

```python
renamed_df = reduced_df.rename(columns={ "2015": "2015 Total GDP", "2016": "2016 Total GDP", "2017": "2017 Total GDP", "2018": "2018 Total GDP", "2019": "2019 Total GDP",})
renamed_df.head()
```

```python
renamed_df.count()
```

```python
renamed_df.to_csv("Output/GDP_API_clean.csv")
```
