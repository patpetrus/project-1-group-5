## File Merge

```python
# Set file paths
world_happiness_2015 = "Output/2015_clean.csv"
world_happiness_2016 = "Output/2016_clean.csv"
world_happiness_2017 = "Output/2017_clean.csv"
world_happiness_2018 = "Output/2018_clean.csv"
world_happiness_2019 = "Output/2019_clean.csv"
```

```python
# Import 2015 into Data Frame
whs_2015_df = pd.read_csv(world_happiness_2015)
whs_2015_df.head()
```

```python
# Import 2016 into Data Frame
whs_2016_df = pd.read_csv(world_happiness_2016)
whs_2016_df.head()
```

```python
# Import 2017 into Data Frame
whs_2017_df = pd.read_csv(world_happiness_2017)
whs_2017_df.head()
```

```python
# Import 2018 into Data Frame
whs_2018_df = pd.read_csv(world_happiness_2018)
whs_2018_df.head()
```

```python
# Import 2019 into Data Frame
whs_2019_df = pd.read_csv(world_happiness_2019)
whs_2019_df.head()
```

```python
# Merge 2015 & 2016
whs_merge_1 = whs_2015_df.merge(whs_2016_df, how='left', on='Country Name')
whs_merge_1.head()
```

```python
# Merge in 2017
whs_merge_2 = whs_merge_1.merge(whs_2017_df, how='left', on='Country Name')
whs_merge_2.head()
```

```python
# Merge in 2018
whs_merge_3 = whs_merge_2.merge(whs_2018_df, how='left', on='Country Name')
whs_merge_3.head()
```

```python
# Merge in 2019
whs_combined_df = whs_merge_3.merge(whs_2019_df, how='left', on='Country Name')
whs_combined_df.head()
```

```python
# Sort by Country Name
whs_combined_df = whs_combined_df.sort_values(by=['Country Name'])
whs_combined_df.head()
```

```python
# Print dataframe to CSV file
whs_combined_df.to_csv('Output/whs_combined.csv', index=False)
```
