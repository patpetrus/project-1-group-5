# Conclusion

Upon analysis of the data from the World Bank Group and the World Happiness Report, we take the following takeaways:

1. Many of the expected data that we pulled from the World Bank Data set did not result in a correlation; nulling our hypothesis that population, unemployment, GDP, military expenditure, and labor force. However, the result of these null hypotheses is not a result of extreme outliers in the data sets skewing the results.
2. Data sets that showed a moderate positive correlation to the countries' happiness score are a country's population of adults and electricity access. The data set that had a moderate negative correlation to a country's happiness is infant mortality. From this analysis, access to electricity's moderate positive correlation is skewed; as we enter the 2020's, more countries have advanced to the point that majority of the population has access to electricity. Making it clear that access to electricity is part of a nation's core infrastructure, so having widely available electricity is indicative (though not directly tied to) a country's also having widely available water treatment, sewage, roadworks, etc. Countries without electricity may find it more difficult to build these other core infrastructure systems, leading to a negative feedback loop, which manifests in, among other things, lower overall happiness scores.
3. The two data sets that show a strong positive correlation are female and male life expectancy. From this and the moderate negative correlation of infant mortality, you can conclude that we analyzed health from the data that health is the critical factor for a country's happiness.

Though we have concluded that health is the key contributing factor to a country's happiness score, additional analyses can be done for further research.

1. While looking into contributing factors for a country's happiness score, many of the data sets that initially would be good predictors of happiness had incomplete country data. This caused the data to be unusable for this initial project. Further analysis of the data is needed, and factors need to be considered when pulling incomplete data. With insufficient data sets, looking at the two inner quartiles might prove useful to limit the risk of extreme outliers of incomplete data. Another is to take ten countries in each quartile in the happiness score index and use them as a baseline for incomplete data. But ultimately would be used as supporting analysis, not the primary for determination of a countries happiness. 
2. A year-to-year comparison of the complete data for a country to see if happiness scores go up as countries develop, as factors improve/decrease, and does the happiness score correlates to the increases and decreases.
3. There is an endless supply of data sets from the World Bank Group that could further analyze the world happiness score. This analysis is just an initial start of a more in-depth look into contributing factors to a countries happiness.
