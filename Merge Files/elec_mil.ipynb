```python
# Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
from scipy.stats import sem
from scipy.stats import linregress
```

```python
# Set path and read CSV
path = 'MERGE.csv'
merged_df = pd.read_csv(path)
merged_df
```

```python
# Military Expenditure vs. World Happiness
```

```python
# 2015 Total % of Military Expenditure and 2015 World Happiness
plt.scatter(merged_df['2015 Happiness Score'], merged_df['2015 Military Expenditure (%GDP)'])
plt.ylabel('2015 Military Expenditure (%GDP)')
plt.x2015 Happiness Scorelabel('2015 Happiness Score')
plt.title('2015 Happiness Scores and Military Expenditure (%GDP)')
plt.show()
```

```python
# 2016 Total % of Military Expenditure and 2016 World Happiness
plt.scatter(merged_df['2016 Happiness Score'], merged_df['2016 Military Expenditure (%GDP)'])
plt.ylabel('2016 Military Expenditure (%GDP)')
plt.xlabel('2016 Happiness Score')
plt.title('2016 Happiness Scores and Military Expenditure (%GDP)')
plt.show()
```

```python
# 2017 Total % of Military Expenditure and 2017 World Happiness
plt.scatter(merged_df['2017 Happiness Score'], merged_df['2017 Military Expenditure (%GDP)'])
plt.ylabel('2017 Military Expenditure (%GDP)')
plt.xlabel('2017 Happiness Score')
plt.title('2017 Happiness Scores and Military Expenditure (%GDP)')
plt.show()
```

```python
# 2018 Total % of Military Expenditure and 2018 World Happiness
#note the space before '2018 military expenditure'
plt.scatter(merged_df['2018 Happiness Score'], merged_df[' 2018 Military Expenditure (%GDP)'])
plt.ylabel('2018 Military Expenditure (%GDP)')
plt.xlabel('2018 Happiness Score')
plt.title('2018 Happiness Scores and Military Expenditure (%GDP)')
plt.show()
```

```python
# 2019 Total % of Military Expenditure and 2019 World Happiness
plt.scatter(merged_df['2019 Happiness Score'], merged_df['2019 Military Expenditure (%GDP)'])
plt.ylabel('2019 Military Expenditure (%GDP)')
plt.xlabel('2019 Happiness Score')
plt.title('2019 Happiness Scores and Military Expenditure (%GDP)')
plt.show()
```

```python
# 2015 Total % of Military Expenditure and 2015 World Happiness

military_2015 = merged_df.iloc[:, 44]
whs_2015 = merged_df.iloc[:, 58]
mil_whs_2015 = st.pearsonr(military_2015, whs_2015)
print(f'The correlation between total military expenditure and world happiness score in 2015 is {round(mil_whs_2015[0],2)}.')

x_adults_2015 = merged_df['2015 Happiness Score']
y_adults_2015 = merged_df['2015 Military Expenditure (%GDP)']
slope, intercept, rvalue, pvalue, stderr = linregress(x_adults_2015, y_adults_2015)
regressValues = x_adults_2015 * slope + intercept
lineEq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

plt.scatter(x_adults_2015, y_adults_2015)
plt.plot(x_adults_2015, regressValues,"r-")
plt.annotate(lineEq,(6,10),fontsize=15,color="red")
plt.xlabel('2015 Happiness Score')
plt.ylabel('2015 Military Expenditure (%GDP)')
plt.title('2015 Happiness Scores and Total Military Expenditure (%GDP)')
plt.show()
```
