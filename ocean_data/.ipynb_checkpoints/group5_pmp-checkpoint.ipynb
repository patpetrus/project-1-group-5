```python
# Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
from scipy.stats import sem
from scipy.stats import linregress
```

```python
# Set path and read CSV
path = '../ocean_data/MERGE.csv'
merged_df = pd.read_csv(path)
merged_df
```

## Infant Mortality Rate and World Happiness Score

```python
# 2015 Infant Mortality Rate and 2015 World Happiness

imr_2015 = merged_df.iloc[:, 30]
whs_2015 = merged_df.iloc[:, 58]
imr_whs_2015 = st.pearsonr(imr_2015, whs_2015)
print(f'The correlation between infant mortality and world happiness score in 2015 is {round(imr_whs_2015[0],2)}.')

x_imr_2015 = merged_df['2015 Happiness Score']
y_imr_2015 = merged_df['2015 Mortality (<5)']
slope, intercept, rvalue, pvalue, stderr = linregress(x_imr_2015, y_imr_2015)
regressValues = x_imr_2015 * slope + intercept
lineEq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

plt.scatter(x_imr_2015, y_imr_2015)
plt.plot(x_imr_2015, regressValues,"r-")
plt.annotate(lineEq,(6,10),fontsize=15,color="red")
plt.xlabel('2015 Happiness Score')
plt.ylabel('2015 Infant Mortality Rate (per 1k)')
plt.title('2015 Happiness Scores and Infant Mortality Rate')
plt.show()
```
