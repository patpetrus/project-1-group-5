```python
# Dependencies
import pandas as pd

# Set up file paths
pop_unemployment_path = '../ocean_data/pop_unemployment.csv'
gdp_path = '../project_data_teare/Output/GDP_API_clean.csv'
labor_path = '../project_data_teare/Output/Labor_API_clean.csv'
gov_expend_path = '../project_data_petrus/ed_spending.csv'
```

```python
# Import population and unemployment CSVs
pop_unemployment_df = pd.read_csv(pop_unemployment_path)
pop_unemployment_df = pop_unemployment_df.drop(columns=['Unnamed: 0'])
pop_unemployment_df.head()
```

```python
# Import GDP CSV
gdp_df = pd.read_csv(gdp_path)
gdp_df = gdp_df.drop(columns=['Unnamed: 0'])
# gdp_df = gdp_df.drop(columns=['Country Code'])
# gdp_df = gdp_df.drop(columns=['Indicator Name'])
# gdp_df = gdp_df.rename(columns={'2015': '2015 GDP', '2016': '2016 GDP', '2017':'2017 GDP', '2018': '2018 GDP', 
#                                 '2019': '2019 GDP'})
gdp_df.head()
```

```python
# Import Labor CSV
labor_df = pd.read_csv(labor_path)
labor_df = labor_df.drop(columns=['Unnamed: 0'])
# labor_df = labor_df.drop(columns=['Country Code'])
# labor_df = labor_df.drop(columns=['Indicator Name'])
# labor_df = labor_df.rename(columns={'2015': '2015 Total Labor Force', '2016': '2016 Total Labor Force', 
#                                     '2017':'2017 Total Labor Force', '2018': '2018 Total Labor Force', 
#                                     '2019': '2019 Total Labor Force'})
labor_df.head()
```

```python
# Import government expenditure 
gov_expend_df = pd.read_csv(gov_expend_path)
gov_expend_df = gov_expend_df.drop(columns=['Unnamed: 0'])
gov_expend_df.head()
```

```python
# Merge pop/unemployment and GDP Dataframes
pop_unemployment_df = pop_unemployment_df.merge(gdp_df, how='left', on='Country Name')
pop_unemployment_df.head()
```

```python
# Merge pop/unemployment/GDP and labor Dataframes
pop_unemployment_df = pop_unemployment_df.merge(labor_df, how='left', on='Country Name')
pop_unemployment_df.head()
```

```python
# # Merge pop/etc and gov expend Datafrans
pop_unemployment_df = pop_unemployment_df.merge(gov_expend_df, how='left', on='Country Name')
pop_unemployment_df.head()
```
