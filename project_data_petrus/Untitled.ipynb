```python
import pandas as pd

total_gov_expenditure = "total_gov_expenditure.csv"
total_gov_expenditure_df = pd.read_csv(total_gov_expenditure)
total_gov_expenditure_df = total_gov_expenditure_df.rename(columns={'2015 [YR2015]': 'Total 2015', '2016 [YR2016]': 'Total 2016', '2017 [YR2017]': 'Total 2017', '2018 [YR2018]': 'Total 2018', '2019 [YR2019]': 'Total 2019'})
total_gov_expenditure_df = total_gov_expenditure_df[['Country Name', 'Total 2015', 'Total 2016', 'Total 2017', 'Total 2018', 'Total 2019']]
total_gov_expenditure_df = total_gov_expenditure_df.drop([264, 265, 266, 267, 268])

total_gov_expenditure_df.head()
```

```python
total_gov_expenditure_df['Total 2015'] = pd.to_numeric(total_gov_expenditure_df['Total 2015'], errors='coerce')
total_gov_expenditure_df['Total 2016'] = pd.to_numeric(total_gov_expenditure_df['Total 2016'], errors='coerce')
total_gov_expenditure_df['Total 2017'] = pd.to_numeric(total_gov_expenditure_df['Total 2017'], errors='coerce')
total_gov_expenditure_df['Total 2018'] = pd.to_numeric(total_gov_expenditure_df['Total 2018'], errors='coerce')
total_gov_expenditure_df['Total 2019'] = pd.to_numeric(total_gov_expenditure_df['Total 2019'], errors='coerce')
total_gov_expenditure_df.dtypes
```

```python
total_gov_expenditure_df.round(3)
```
