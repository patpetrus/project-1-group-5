```python
import pandas as pd
```

```python
labor_file = "Resources/Labor_API.csv"
```

```python
labor_file_df = pd.read_csv(labor_file, skiprows=4)
labor_file_df.head()
```

```python
labor_file_df.columns
```

```python
reduced_df = labor_file_df[[  'Country Name', '2015', '2016', '2017', '2018', '2019']]
reduced_df.head()
```

```python
renamed_df = reduced_df.rename(columns={'2015':'2015 Total Labor Force', '2016':'2016 Total Labor Force', '2017': '2017 Total Labor Force', '2018':'2018 Total Labor Force', '2019': '2019 Total Labor Force'})
renamed_df.head()
```

```python
renamed_df.count()
```

```python
renamed_df.to_csv("Output/Labor_API_clean.csv")
```
