```python
import pandas as pd
```

```python
gdp_file = "Resources/GDP_API.csv"
```

```python
gdp_file_df = pd.read_csv(gdp_file)
gdp_file_df.head()
```

```python
new_header = gdp_file_df.iloc[3]
gdp_file_df = gdp_file_df[4:]
gdp_file_df.columns = new_header
gdp_file_df.head(10)
```

```python
gdp_file_df.columns
```

```python
reduced_df = gdp_file_df[[  'Country Name',   'Country Code', 'Indicator Name',                 2015.0,
                 2016.0,           2017.0,           2018.0,           2019.0]]
reduced_df.head()
```

```python
renamed_df = reduced_df.rename(columns={                   2015.0: 2015,
                 2016.0: 2016,           2017.0: 2017,           2018.0: 2018,           2019.0: 2019,})
renamed_df.head()
```

```python
renamed_df.count()
```

```python
renamed_df.round(decimals=0)
```

```python
renamed_df.to_csv("Output/GDP_API_clean.csv")
```
