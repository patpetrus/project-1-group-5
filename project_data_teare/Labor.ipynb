```python
import pandas as pd
```

```python
labor_file = "Resources/Labor_API.csv"
```

```python
labor_file_df = pd.read_csv(labor_file)
labor_file_df.head()
```

```python
new_header = labor_file_df.iloc[3]
labor_file_df = labor_file_df[4:]
labor_file_df.columns = new_header
labor_file_df.head()
```

```python
labor_file_df.columns
```

```python
reduced_df = labor_file_df[[  'Country Name',   'Country Code', 'Indicator Name',                 2015.0,
                 2016.0,           2017.0,           2018.0,           2019.0]]
reduced_df.head()
```

```python
renamed_df = reduced_df.rename(columns={                   2015.0: 2015,
                 2016.0: 2016,           2017.0: 2017,           2018.0: 2018,           2019.0: 2019,})
renamed_df.head()
```

```python
renamed_df.count()
```

```python
renamed_df.round(decimals=0)
```

```python
renamed_df.to_csv("Output/Labor_API_clean.csv")
```
